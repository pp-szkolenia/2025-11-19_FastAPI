## Tryby komórki – edycji i poleceń

Każda komórka (zarówno kod jak i markdown) może być w **jednym z dwóch trybów**:
- trybie edycji
- trybie poleceń

Aby znaleźć się w **trybie edycji**, musimy *wejść* do komórki (`Enter`). Wtedy możemy edytować jej zawartość, np. modyfikować kod czy tworzyć opis w markdown.

Aby znaleźć się w **trybie poleceń**, musimy *wyjść* z komórki (`Escape`). Wtedy możemy korzystać z **komend**, np. tworzyć czy usuwać komórki.

**Poniżej kilka przydatnych informacji:**
- Zwróć uwagę na zmianę koloru tła komórki po przełączeniu się między tymi trybami
- Wykonywanie komórek jest możliwe w obu trybach
- Jeśli wejdziesz do komórki z markdown, pamiętaj że aby znowu **zobaczyć podgląd markdown**, musisz tę komórkę uruchomić! Bez tego nadal będziesz mieć **surowy widok**

> Przełącz się między trybem edycji i trybem poleceń. W razie potrzeby, wykonaj komórkę z markdown odpowiednim skrótem klawiszowym

## Tworzenie i usuwanie komórek

Poniższe instrukcje działają wyłącznie w **trybie poleceń**

---
- aby utworzyć nową komórkę **powyżej** wciśnij klawisz `a` (***a**bove*)
- aby utworzyć nową komórkę **poniżej** wciśnij klawisz `b` (***b**elow*)
- aby **usunąć komórkę** wciśnij klawisz `x` lub `d d`
- aby **cofnąć** jedną z powyższych operacji wciśnij klawisz `z`

> Dodaj komórki a następnie napisz i wykonaj kod w nich. Później usuń je i cofnij to usunięcie.

## Komórki z kodem i tekstem

Każda nowo utworzona komórka domyślnie będzie zawierać kod

Poniższe instrukcje działają wyłącznie w **trybie poleceń**

---

- żeby zmienić zawartość komórki na **markdown** należy wcisnąć klawisz `m`
- żeby z powrotem zmienić zawartość komórki na **kod** należy wcisnąć klawisz `y`

> Stwórz nową komórkę a następnie zamień ją na markdown. Napisz jakiś tekst i zobacz jak wygląda podgląd. Następnie zamień ją na komórkę z kodem i wykonaj. Na koniec usuń.

- każda komórka może występować w dwóch trybach – **edycji** oraz **poleceń**

- komórki możemy dodać oraz usuwać

**Skróty klawiszowe:**

```
↓ – przejdź do komórki poniżej
↑ – przejdź do komórki powyżej

Shift + Enter – uruchom komórkę i przejdź do następnej
Ctrl + Enter – uruchom komórkę i pozostań w niej
Alt + Enter – uruchom komórkę i utwórz nową poniżej

Enter – wejdź do trybu edycji
Escape – wyjdź do trybu poleceń

-- (poniższe tylko w trybie poleceń) --

a – utwórz komórkę powyżej
b – utwórz komórkę poniżej
x – usuń komórkę
z – cofnij operację

m – zamień komórkę na markdown
y – zamień komórkę na kod
```

## Bonus: tips & tricks

Filozofia pisania kodu w notebookach jest inna, niż w typowym IDE. Dlatego warto stosować się do pewnych wskazówek, które przyspieszą pracę z Jupyterze i sprawią, że będzie ona przyjemniejsza.

**1. Najpierw sprawdź, potem przypisz do zmiennej**

Kiedy zamierzasz przypisać do zmiennej jakieś złożone wyrażenie, którego poprawności nie jesteś pewien, **nie zaczynaj od tworzenia zmiennej** (chociaż to naturalny odruch).

Zamiast tego zrób tak:
- napisz wyrażenie (lub jego część)
- uruchom komórkę (`Ctrl+Enter` by w niej pozostać)
- sprawdź czy wszystko się zgadza
- jeśli tak, wejdź do komórki i przypisz wyrażenie do zmiennej
- jeśli nie, zmodyfikuj je i sprawdź jeszcze raz

> Łatwiej zorientujesz się w kodzie jeżeli będzie go mniej, a natychmiastowe przypisanie wyrażenia do zmiennej w Jupyterze nie jest konieczne

**2. Dobieraj odpowiedni skrót kiedy uruchamiasz komórkę**

- jeżeli wiesz, że po uruchomieniu komórki chcesz dalej ją edytować – `Ctrl + Enter`
- jeżeli wiesz, że po uruchomieniu komórki chcesz dalej przejść do następnej – `Shift + Enter`
- jeżeli wiesz, że po uruchomieniu komórki chcesz od razu utworzyć nową  – `Alt + Enter`

**3. Nie pisz wszystkiego w jednej komórce**

Jeżeli regularnie korzystasz z IDE, naturalne będzie pisanie całego kodu w jednej komórce. Taki sposób jednak **nie zawsze jest optymalny**. Często wygodniej jest podzielić problem na mniejsze etapy i każdy zapisać w osobnej komórce, po drodze sprawdzając wynik każdego etapu. To trochę jak **używanie debuggera** w trakcie pisania kodu.

**4. Wklejaj obrazki do notebooka**

Jedną z zalet notebooków jest to, że możesz wklejać do nich obrazy. Po prostu skopiuj jakiś obraz (albo zrób screenshota) i, będąc w komórce **markdown**, użyj skrótu klawiszwego `Ctrl + V`.
